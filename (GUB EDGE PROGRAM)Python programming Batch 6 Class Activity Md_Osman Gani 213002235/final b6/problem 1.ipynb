```python
import pandas as pd
import numpy as np
```

```python
#ans:1
#1 task
df = pd.read_csv('/content/employees_large.csv')
df.head()
```

```python
#ans:1
#2 task
avg_sal = df.groupby('Department')['Salary'].mean()
print("\nAverage Salary for Department:\n", avg_sal)
```

```python
#ans:1
#task 3
havg_sal = avg_sal.idxmax()
print("Highest Average Salary:", havg_sal)
```
