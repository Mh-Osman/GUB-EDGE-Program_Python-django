```python
import numpy as np

# Define matrices
A = np.array([[2, 4, 5], [1, 3, 2], [3, 6, 1]])
B = np.array([[3, 2, 1], [6, 3, 0], [5, 3, 2]])


```

```python
#t1
sum_matrix = A + B  # 1. Compute the element-wise sum (A + B)
sum_matrix = A + B
print("Element-wise Sum:\n", sum_matrix)

# 2. Compute the element-wise product (A * B)


# 3. Compute the dot product A * B^T (B Transpose)


# 4. Compute the determinant of A

print("Element-wise Sum:\n", sum_matrix)
```

```python
#t2
product_matrix = A * B
print("\nElement-wise Product:\n", product_matrix)
```

```python
#t3
dot_product = np.dot(A, B.T)
print("\nDot Product (A * B^T):\n", dot_product)
```

```python
#t4
det_A = np.linalg.det(A)
print("\nDeterminant of Matrix A:", det_A)
```
