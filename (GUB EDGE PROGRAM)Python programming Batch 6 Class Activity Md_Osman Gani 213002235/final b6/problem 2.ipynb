```python
#ans 2:
#t1
df2=pd.read_csv('/content/sales_large.csv')
df2.head()
```

```python
#ans2
#t2
total_sal = df2.groupby(['Product', 'Region'])['Sales'].sum()
print("\nTotal Sales for each Product in Region:\n", total_sal )
```

```python
#ans2
#t3
highest_sales_product = df2.groupby('Product')['Sales'].sum().idxmax()
print("\nProduct with Highest Total Sales:", highest_sales_product)
```
