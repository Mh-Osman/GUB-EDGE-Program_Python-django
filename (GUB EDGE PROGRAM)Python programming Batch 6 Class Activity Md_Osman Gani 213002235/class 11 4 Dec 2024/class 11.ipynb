```python

class Student:

    def __init__(self, s_id ,name, marks):
        self.s_id = s_id
        self.name = name
        self.marks = marks


    def display(self):
        print("ID:", self.s_id)
        print("Name:", self.name)
        print("Marks:", self.marks)


s1 = Student(101, "John Doe", 85)

s1.display()
```

```python
class Marks:
    def __init__(self):
        self.marks_list = []

    def get_marks(self, number_of_subjects):
        for i in range(number_of_subjects):
            mark = float(input(f"Enter marks for subject {i+1}: "))
            self.marks_list.append(mark)

    def display_total_and_average(self):
        total_marks = sum(self.marks_list)
        average_marks = total_marks / len(self.marks_list) if self.marks_list else 0
        print("Total Marks:", total_marks)
        print("avg Marks:", average_marks)

student_marks = Marks()

num_subjects = int(input("Enter the number of subjects: "))
student_marks.get_marks(num_subjects)

student_marks.display_total_and_average()
```
